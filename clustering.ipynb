```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.cluster import AgglomerativeClustering
from sklearn.model_selection import StratifiedKFold
train_df = pd.read_csv('./iBeacon_RSSI_Unlabeled.csv')
train_df = train_df.drop(['location','date'], axis=1)
```

```python
#DBScan
clustering = DBSCAN(eps=3, min_samples=2).fit(train_df)
clustering=clustering.labels_
```

```python
X_tsne = TSNE(n_components=2,learning_rate=100).fit_transform(train_df)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=clustering)
plt.colorbar()
plt.show() 
```

```python
#Kmeans
kmeans = KMeans(n_clusters=13, random_state=0).fit(train_df)
kmeans=kmeans.labels_
```

```python
X_tsne = TSNE(n_components=2,learning_rate=100).fit_transform(train_df)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=kmeans)
plt.colorbar()
plt.show() 
```

```python
#GMM
np.random.seed(1)
g = GaussianMixture(n_components=13)
obs = np.concatenate((np.random.randn(100, 1),10 + np.random.randn(300, 1)))
g.fit(train_df) 
labels = g.predict(train_df)
```

```python
X_tsne = TSNE(n_components=2,learning_rate=100).fit_transform(train_df)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels)
plt.colorbar()
plt.show() 
```

```python
#AgglomerativeClustering
Agg = AgglomerativeClustering(n_clusters=13).fit(train_df)
Agg = Agg.labels_
```

```python
X_tsne = TSNE(n_components=2,learning_rate=100).fit_transform(train_df)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=Agg)
plt.colorbar()
plt.show() 
```
